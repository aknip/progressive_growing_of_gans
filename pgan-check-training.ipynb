```python
import matplotlib.pyplot as plt
import glob
image_list = []
for filename in glob.glob("./results/001-pgan-celebahq100-preset-v2-1gpu-fp32/fakes*.png"):
    image_list.append(filename)
image_list.sort()
plt.figure(figsize=(20,10)) #20,10
columns = 5 #5
for i, image_name in enumerate(image_list):
    image = plt.imread(image_name)
    plt.subplot(len(image_list) / columns + 1, columns, i + 1)
    plt.imshow(image)
```
